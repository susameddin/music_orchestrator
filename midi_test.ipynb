```python
import os
import mido
import pretty_midi
import music21
```

```python
slakh_path = "/engram/naplab/shared/Slakh2100/slakh2100_flac_redux"
track_path = "test/Track01881"
```

```python
#midi_file = mido.MidiFile(os.path.join(slakh_path,track_path,'MIDI','S03.mid'))
midi_file = mido.MidiFile(os.path.join(slakh_path,track_path,'all_src.mid'))
```

```python
for message in midi_file:
    print(message)
```

```python
for track in midi_file.tracks:
    print(f"Track: {track.name}")
    for msg in track:
        print(msg)
```

```python
midi_data = pretty_midi.PrettyMIDI(os.path.join(slakh_path,track_path,'all_src.mid'))

# List all instruments
for instrument in midi_data.instruments:
    print(f"Instrument: {instrument.name}")

# Access notes for the first instrument
for note in midi_data.instruments[0].notes:
    print(f"Note: {note.pitch}, Start: {note.start}, End: {note.end}")
```

```python
# Get tempo
tempo = midi_data.estimate_tempo()
print(f"Estimated Tempo: {tempo}")

# Get time signature changes
time_signatures = midi_data.time_signature_changes
print(f"Time Signatures: {time_signatures}")

# Get key signature changes
key_signatures = midi_data.key_signature_changes
print(f"Key Signatures: {key_signatures}")
```

```python
from music21 import converter, key

# Load the MIDI file
midi = converter.parse(os.path.join(slakh_path,track_path,'all_src.mid'))

# Analyze the key
key_signature = midi.analyze('key')
print(f"Estimated Key: {key_signature}")
```
