```python
import os
import yaml
import json
import librosa
```

```python
slakh_path = "/engram/naplab/shared/Slakh2100/slakh2100_flac_redux"
```

```python
#Training Set
jsonl = []
for track_id in os.listdir(os.path.join(slakh_path,'train')):
    print(track_id)
    track_path = os.path.join("train",track_id)
    
    all_instruments = []    
    
    audio, sr = librosa.load(os.path.join(slakh_path,track_path,'mix.flac'), sr=None)
    audio_len = len(audio)/sr
    
    with open(os.path.join(slakh_path,track_path,'metadata.yaml'), 'r') as metadata_file:
        metadata = yaml.safe_load(metadata_file) 
        
    for stem in metadata['stems'].keys():
        output_path = os.path.join(slakh_path,track_path,'stems')
        instrument = metadata['stems'][stem]['midi_program_name']
#         curr_json = {"key": "", "artist": "", "sample_rate": sr, "file_extension": "flac", "description": instrument, "keywords": instrument, "duration": round(audio_len,1), "bpm": "", "genre": "", "title": "", "name": "", "instrument": instrument, "moods": ""}
#         with open(os.path.join(slakh_path,track_path,'stems',stem+'.json'), 'w') as json_file:
#             json.dump(curr_json, json_file, indent=4)
#         jsonl.append({"path": os.path.join(slakh_path,track_path,'stems',stem+'.flac'), "duration": audio_len, "sample_rate": sr, "amplitude": None, "weight": None, "info_path": None})
        all_instruments.append(instrument)    
        
    all_instruments = ', '.join(all_instruments)
    output_path = os.path.join(slakh_path,track_path)
#     curr_json = {"key": "", "artist": "", "sample_rate": sr, "file_extension": "flac", "description": all_instruments, "keywords": all_instruments, "duration": round(audio_len,1), "bpm": "", "genre": "", "title": "", "name": "", "instrument": all_instruments, "moods": ""}
#     with open(os.path.join(slakh_path,track_path,'mix.json'), 'w') as json_file:
#         json.dump(curr_json, json_file, indent=4)
    jsonl.append({"path": os.path.join(slakh_path,track_path,'mix.flac'), "duration": audio_len, "sample_rate": sr, "amplitude": None, "weight": None, "info_path": None})

    
jsonl_path = "/home/sd3705/music_gen_2024f/audiocraft/egs/slakh2100_only_mix_train"
if not os.path.exists(jsonl_path):
    os.makedirs(jsonl_path)
with open(os.path.join(jsonl_path,"data.jsonl"), 'w') as jsonl_file:
    for curr_data in jsonl:
        jsonl_file.write(json.dumps(curr_data) + '\n')    
```
